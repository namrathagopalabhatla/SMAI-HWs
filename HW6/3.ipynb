```python
import numpy as np
from numpy import sqrt
import matplotlib.pyplot as plt
from sympy import Symbol
from sympy.solvers import solve
```

```python
w1 = np.array([[0, 0], [0, 1], [2, 0], [3, 2], [3, 3], [2, 2],[2, 0]])

w2 = np.array([[7, 7], [8, 6], [9, 7],[8, 10],[7, 10],[8, 9],[7, 11]])
x1 = Symbol('x1')
x2 = Symbol('x2')

solution_d = solve([(1.231 * (x1 ** 2)) - (0.6727 * (x2 ** 2)) + (1.9188 * (x1 * x2)) - (40.3801 * x1) - (11.5247 * x2) + 204.7203])
solution_e = solve([0.5 * ((1.231 * (x1 ** 2)) - (0.6727 * (x2 ** 2)) + (1.9188 * (x1 * x2)) - (40.3801 * x1) - (11.5247 * x2) + 204.7203) + np.log(2)])

print('(d) Equation: x1 = {}'.format(solution_d[0][x1]))
print('(e) Equation: x1 = {}'.format(solution_e[0][x1]))

x2 = np.linspace(0, 10, 1000)

plt.figure(figsize=(15, 15))
plt.scatter(w1[:, 0], w1[:, 1], c='b', label='Points in W1')
plt.scatter(w2[:, 0], w2[:, 1], c='r', label='Points in W2')
plt.plot(-0.779366368805849*x2 - 4.06173842404549e-5*sqrt(699416824.0*x2**2 - 9821504896.0*x2 + 62250971881.0) + 16.4013403736799, x2, label='Decision Boundary', c='green')
plt.plot(-0.779366368805849*x2 - 4.06173842404549e-5*sqrt(699416824.0*x2**2 - 9821504896.0*x2 + 61568505481.0) + 16.4013403736799, x2, label='New Decision Boundary', c='orange')
plt.legend()
plt.show()
```
